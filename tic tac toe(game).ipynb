```python
# -*- coding: utf-8 -*-
"""
Created on Wed Mar 11 23:55:59 2020

@author: manish
"""

from IPython.display import clear_output

dict_place = {} #dictionary to hold the X/O mark corresponding to the place
dict_player ={} #dict  to keep details of player 1 and player 2
place = ["place_" + str(x) for x in range(1,10)]


# function to initalize every field of boared with " "
def dict_initialize():
    for i in place:
        dict_place.update({i: " "})

        
#fuc to print welcome message and take player details
def welcome():
    print("\n\n~~~~~~~~~~~~~~~~~~~~ Welcome To 'TIC-TAC-TOE' Game ~~~~~~~~~~~~~~~~~~~~\n\n")
    player1 = input("Enter first player name : ").capitalize()
    player2 = input("Enter second player name : ").capitalize()
    dict_player.update({"player1" : [player1, 0]})
    dict_player.update({"player2" : [player2, 0]})
    print("\n~~~~~~~~~~~~~~~~~~~~~~~~~ Start Playing ~~~~~~~~~~~~~~~~~~~~~~~~~")
    
    
#function to print the tic-tac-toe board 
def print_board():
    clear_output()
    #first row
    print(f"\n\t\t {dict_place['place_7']} | {dict_place['place_8']} | {dict_place['place_9']}  ") 
    print(f"\t\t-----------")
    #second row
    print(f"\t\t {dict_place['place_4']} | {dict_place['place_5']} | {dict_place['place_6']}  ")
    print(f"\t\t-----------")
    #third row
    print(f"\t\t {dict_place['place_1']} | {dict_place['place_2']} | {dict_place['place_3']}  ")
  
          
#To print current score of the players
def print_score():
    print(f" {dict_player['player1'][0]} : {dict_player['player1'][1]}")
    print(f" {dict_player['player2'][0]} : {dict_player['player2'][1]}")

#function to check in rows for the three continious mark                   
def check_row():
    flag = False
#     print("in row")
    #row one
    if(dict_place['place_7'] == dict_place['place_8'] == dict_place['place_9']):#if all are same in first row
        #check if all isn't null
        if not ((dict_place['place_7'] == " ") and (dict_place['place_8'] == " ") and (dict_place['place_9'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_7']} has been found at : row 1")
            
    if(dict_place['place_4'] == dict_place['place_5'] == dict_place['place_6']):
        if not((dict_place['place_4'] == " ") and (dict_place['place_5'] == " ") and (dict_place['place_6'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_4']} has been found at : row 2")
          
    if(dict_place['place_1'] == dict_place['place_2'] == dict_place['place_3']):
        if not((dict_place['place_1'] == " ") and (dict_place['place_2'] == " ") and (dict_place['place_3'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_1']} has been found at : row 3")
        
    return flag

#function to check in columns for the three continious mark
def check_column():
    flag = False
#     print("in column")
    if(dict_place['place_7'] == dict_place['place_4'] == dict_place['place_1']):
        if not((dict_place['place_7'] == " ") and (dict_place['place_4'] == " ") and (dict_place['place_1'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_7']} has been found at : column 1")
    if(dict_place['place_8'] == dict_place['place_5'] == dict_place['place_2']):
        if not((dict_place['place_8'] == " ") and (dict_place['place_5'] == " ") and (dict_place['place_2'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_8']} has been found at : column 2")
    if(dict_place['place_9'] == dict_place['place_6'] == dict_place['place_3']):
        if not((dict_place['place_9'] == " ") and (dict_place['place_6'] == " ") and (dict_place['place_3'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_9']} has been found at : column 3")
        
    return flag

#function to check in diagonals for the three continious mark
def check_diagonal():
    flag = False
#     print("in dignl")
    if(dict_place['place_7'] == dict_place['place_5'] == dict_place['place_3']):
        if not((dict_place['place_7'] == " ") and (dict_place['place_5'] == " ") and (dict_place['place_3'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_7']} has been found at : Diagonal 1")
    if(dict_place['place_9'] == dict_place['place_5'] == dict_place['place_1']):
        if not((dict_place['place_9'] == " ") and (dict_place['place_5'] == " ") and (dict_place['place_1'] == " ")):
            flag = True
            print(f"\nThree connecting {dict_place['place_9']} has been found at : Diagonal 2")
        
    return flag

#check for the thee consecutives marks in all three ways(rows,columns, diagonal
def check_for_winner():
    return  (check_row() or check_column() or check_diagonal() )
    
    

#function to take input from the player
def take_input(mark):
    while(1):
        user_input = input(f"Enter the place, to put mark \'{mark}\' : ")
        
        #to check if the user_input is integer only
        if (not user_input.isdigit()):
            print("\nInvalid Input !!!!\nPlease enter only number between '1 - 9' ...")
            print("Try again ........")
            continue
            
        #to check if the value is in digit range(1,10)
        if (int(user_input) not in range(1,10)):
            print("\nInvalid Input !!!!\nEnter the place value only between '1 - 9' ...")
            print("Try again ........")
            continue
            
        place = "place_" + user_input 
        
        #check if place has already some value
        if not (dict_place[place] == " "):
            print(f"\nPlace '{user_input}' already has {dict_place[place]} value !!.")
            print("Try again ........")
            continue
        else:
            break
    
    dict_place[place] = mark #updating the place value in dict/board
    print_board()#print updated board

#funtion to take users input in terms of place( 1 - 9)          
def initiate_input():                
    x_flag = o_flag = False
                  
    for i in range(5):
        print(f"\n{dict_player['player1'][0]},")  #print player1 name        
        take_input("X")
        if(i > 1):
          x_flag = check_for_winner()
          if x_flag: 
            print(f"\n\'{dict_player['player1'][0]}\' has Won the match !!!\n")   
            dict_player['player1'][1] += 1
            break

        #to break last time as input can be taken only nine times   
        if i == 4:
            break
                  
        print(f"\n{dict_player['player2'][0]},") #print player 2 name
        take_input("O")
              
        if(i > 1):
          o_flag = check_for_winner()
          if o_flag:
              print(f"\n\'{dict_player['player2'][0]}\' has Won the match !!!\n")
              dict_player['player2'][1] += 1
              break
                 
    #check for draw match    
    if not (x_flag or o_flag):
        print("\n\n!! This match is draw. Well Played guys. !!")
    #After complition of match print the current score of players              
    print(f"\nCurrent score --------------")
    print_score()
    
#fnction for askinig if players want to continue with game    
def ask_continuation():
    choice = input("Do you guys want to continue challenging each other? (y/n)\n").lower()
    #return false if user doesn't want to continue or else return true
    return not (choice == "n")
       

#function to start the new game each time                  
def start_game():
    dict_initialize()# to intialize dict_place with " " value
    print_board()#print initail board with no value
    initiate_input()#ask input from user 

#program exectuion will strat from here    
welcome()# to print welcme message and get players name
print_score()
start_game()
                  
while(1):
    flag_choice = ask_continuation()
    if(flag_choice):
        print("\nContinuing with next game .............")
        start_game()# if chice is not "n" then start next game
    else:
        print("\n~~~~~~~~!!!!!!! THANKS FOR PLAYING !!!!!!!~~~~~~~~")
        print("Your final score : ")
        print_score()          
        break
    
      
```
